```python
#DeepSlice only accepts .png files
#All our mouse brain slice files are .tif
#This script converts files from .tif to .png



#load packages
import os 
from PIL import Image
```

```python
#convert controls

os.chdir("controls")
os.chdir("cerebellum_4x")
    
for filename in os.listdir():
    if filename.lower().endswith((".tif", ".tiff")):
        img = Image.open(filename)

        png_name = os.path.splitext(filename)[0] + ".png"
        img.save(png_name)

        print(f"Converted {filename} -> {png_name}")
os.chdir(r"..")
os.chdir(r"hippocampus_4x")

for filename in os.listdir():
    if filename.lower().endswith((".tif", ".tiff")):
        img = Image.open(filename)

        png_name = os.path.splitext(filename)[0] + ".png"
        img.save(png_name)

        print(f"Converted {filename} -> {png_name}")

os.chdir(r"..")
os.chdir("midbrain_4x")
for filename in os.listdir():
    if filename.lower().endswith((".tif", ".tiff")):
        img = Image.open(filename)

        png_name = os.path.splitext(filename)[0] + ".png"
        img.save(png_name)

        print(f"Converted {filename} -> {png_name}")
os.chdir(r"..")
os.chdir("septum_4x")
for filename in os.listdir():
    if filename.lower().endswith((".tif", ".tiff")):
        img = Image.open(filename)

        png_name = os.path.splitext(filename)[0] + ".png"
        img.save(png_name)

        print(f"Converted {filename} -> {png_name}")
```
